2. Following data shows the change in population of different animals from 2017 to 2022.

a. Create a pie chart for the population of each year

b. Can we show all the pie chart on the single pane? Write the code.


```python
import matplotlib.pyplot as plt
years = [2017,2018,2019,2020,2021,2022]
animals = ['bears','dolphins','whales']
bears = [8,54,93,116,137,184]
dolphins = [150,77,32,11,6,1]
whales = [80,54,100,76,93,72]
total = [(8+150+80),(54+77+54),(93+32+100),(116+11+76),(137+6+93),(184+1+92)]
plt.subplot(2,2,1)
plt.title('Total')
plt.pie(total,labels = years)
plt.subplot(2,2,2)
plt.title('Bears')
plt.pie(bears,labels=years)
plt.subplot(2,2,3)
plt.title('Dolphins')
plt.pie(dolphins,labels=years)
plt.subplot(2,2,4)
plt.title('Whales')
plt.pie(whales, labels= years)
```
