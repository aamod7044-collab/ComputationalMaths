12. Create three vectors in 3D space and plot them using Matplotlib.

```python
import matplotlib.pyplot as plt
import numpy as np

U= np.array([1, 0.25 ,0])
V = np.array([0, 1, 0.25])
W = np.array([0.25, 0, 1])

fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')

ax.quiver(0,0,0,U[0],U[1],U[2],color='red')
ax.quiver(0,0,0,V[0],V[1],V[2],color='blue')
ax.quiver(0,0,0,W[0],W[1],W[2],color='green')
```

13.Create two vectors in 2D space, and add them to create a third vector, and plot all three vectors

```python
import matplotlib.pyplot as plt
import numpy as np

U = np.array([-0.5, 0.5])
V = np.array([1,1])

W = U + V

fig = plt.figure()
diag = fig.add_subplot(111, projection= '3d')

diag.quiver(0,0,0,U[0],U[1],0,color='red')
diag.quiver(0,0,0,V[0],V[1],0,color='blue')
diag.quiver(0,0,0,W[0],W[1],0,color ='green')
```

15. Write a function to calculate the direction cosines of the vector. For example, let the vector be
O = (1.5, 1.5, 2.0).

```python
import math
O = (1.5, 1.5, 2.0)
mag = math.sqrt((O[0])**2 + (O[1])**2 + (O[2])**2)

x = O[0]/mag
y = O[1]/mag
z = O[2]/mag

print('X Cosine: ',x,'\n Y Cosine: ',y,'\n Z Cosine: ',z)
```

17. Write a function to calculate Vector (cross) product of two vectors. For example, let the vectors
be a=(1.5, 1.5, 2.0) and b=(−1.5, 0.5, 2.0).

```python
import math
import numpy as np

def cross_product(a, b):
    return (
        a[1]*b[2] - a[2]*b[1],
        a[2]*b[0] - a[0]*b[2],
        a[0]*b[1] - a[1]*b[0]
    )

a = (1.5, 1.5, 2.0)
b = (-1.5, 0.5, 2.0)

result = cross_product(a, b)
print("Cross product:", result)
```
