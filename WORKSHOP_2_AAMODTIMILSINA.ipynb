***Exercise 1***: Write a Python function that takes a list and returns a new list with distinct elements from the
first list.
Sample List : [1,2,3,3,3,3,4,5]
Unique List : [1, 2, 3, 4, 5]

```python
def get_unique_elements(input_list):
    # Convert list to a set to remove duplicates
    return list(set(input_list))

# Sample List
sample_list = [1, 2, 3, 3, 3, 3, 4, 5]

# Function Call
unique_list = get_unique_elements(sample_list)
print(unique_list)
```

2. Write a Python function that takes a number as a parameter and checks whether the number
is prime or not

```python
def is_prime(number):
    # Prime numbers must be strictly greater than 1
    if number <= 1:
        return False

    # We only need to check for factors up to the square root of the number
    # If a number has a factor larger than its square root, the other factor
    # would have to be smaller than the square root.
    limit = int(number**0.5) + 1

    for i in range(2, limit):
        # If the number is evenly divisible by 'i', it's not prime
        if number % i == 0:
            return False

    # If we get through the loop without finding any factors, it is prime
    return True

numbers_to_test = [1, 2, 7, 15, 29, 100]

for num in numbers_to_test:
    if is_prime(num):
        print(f"{num} is a prime number.")
    else:
        print(f"{num} is NOT a prime number.")
```

3. Write a program to print twin primes less than 1000. If two consecutive odd numbers are
both prime then they are known as twin primes.

```python
# Function to check if a number is prime
def is_prime(n):
    if n < 2:
        return False
    for i in range(2, int(n**0.5) + 1):
        if n % i == 0:
            return False
    return True

# Find and print twin primes less than 1000
for i in range(3, 1000, 2):  # check only odd numbers
    if is_prime(i) and is_prime(i + 2):
        print(f"({i}, {i+2})")
```

4. A positive integer is called a perfect number if it is equal to the sum of all of its divisors,
including 1 but excluding the number itself. For example, 6 = 1 + 2 + 3. Write a
function sumOfDivisors that takes a positive integer n and returns the sum of all its divisors
(excluding n).

```python
def sumOfDivisors(n):
    total = 0

    for i in range(1, n):
        if n % i == 0:
            total += i

    return total
    print(sumOfDivisors(6))   # Output: 6
print(sumOfDivisors(10))  # Output: 8 (1 + 2 + 5)
```

```python
def sumOfDivisors(n):
    total = 0
    for i in range(1, n):
        if n % i == 0:
            total += i
    return total

# Find next perfect number after 6
num = 7

while True:
    if sumOfDivisors(num) == num:
        print("Next perfect number is:", num)
        break
    num += 1
```

6. Write the python program to find the mean median from the following list:
Marks = [20, 30, 35, 55, 70, 80, 90] Do not use the library function if there is one.

```python
marks = [20, 30, 35, 55, 70, 80, 90]

# Mean
total = 0
for i in marks:
    total += i
mean = total / len(marks)

# Median
n = len(marks)

# (List is already sorted, so no need to sort)
if n % 2 != 0:
    median = marks[n // 2]
else:
    median = (marks[n // 2 - 1] + marks[n // 2]) / 2

print("Mean =", mean)
print("Median =", median)
```

7. Write a function that converts a decimal number to binary number.

```python
def decimalToBinary(n):
    if n == 0:
        return "0"

    binary = ""

    while n > 0:
        remainder = n % 2
        binary = str(remainder) + binary
        n = n // 2

    return binary
    print(decimalToBinary(10))
print(decimalToBinary(25))
```

8. Write a function prodDigits() that inputs a number and returns the product of digits of that
number.

```python
def prodDigits(n):
    product = 1

    while n > 0:
        digit = n % 10
        product *= digit
        n = n // 10

    return product
    print(prodDigits(123))
print(prodDigits(405))
```

9. Write a function that will take in an unknown number of arguments and multiply all of them
together and run the function for these 2 sets of numbers:
a. 1,2,3,4,5
b. 12,13,14

```python
def multiplyAll(*args):
    product = 1
    for num in args:
        product *= num
    return product

# Test cases
print(multiplyAll(1, 2, 3, 4, 5))   # a
print(multiplyAll(12, 13, 14))      # b
```
