8. **Plot COSINE function:**

y=cos(x), −3π≤x≤3π

```python
import matplotlib.pylab as plt
import math

start = (-3*math.pi)
end = (3*math.pi)
i = start
x=[]
y=[]

while (i <= end):
  x.append(i)
  y.append(math.cos(i))
  i += 0.1

plt.grid()
plt.plot(x,y,color='orange')
```

9. **Plot a circle with:**

a = b = 2

r = 4

x = (rcos(𝜃)+a)

𝑦 = (𝑟sin(𝜃)+𝑏) , where 0 ≤ 𝜃 ≤2𝜋


```python
import matplotlib.pylab as plt
import math

a = 2
b = 2
r = 4
start = 0
end = 2*math.pi +1
i = start
x = []
y = []

while i <= end:
  x.append(r*math.cos(i)+a)
  y.append(r*math.sin(i)+b)
  i += 0.1
plt.xlim = (0,100)
plt.ylin = (0,100)
plt.grid()
plt.plot(x,y,color = 'red')
plt.plot(a,b,marker='o')
plt.gca().set_aspect('equal', adjustable='box')
```

10. **Plot the circle and ellipse on same axes**:

x = a cos(θ)+p,

y = b sin(θ) +q


```python
import matplotlib.pylab as plt
import math

a = 3
b = 2
r = 4
p = 4
q = 4
start = 0
end = 2*math.pi + 1
i = start
x = []
y = []
x1 = []
y1 = []
while i <= end:
    x.append(r*math.cos(i)+a)
    y.append(r*math.sin(i)+b)
    x1.append(a*math.cos(i)+p)
    y1.append(b*math.sin(i)+q)
    i += 0.1
plt.grid()
plt.plot(x, y, color='red')
plt.plot(a, b, marker='o')
plt.plot(x1, y1, color='green')
plt.plot(p, q, marker='o')
plt.legend()
plt.gca().set_aspect('equal', adjustable='box')
```

11. **Plot a parabola:**

(𝑦 − 𝑝) = 4𝑎(𝑥 − 𝑞)^2

```python
import matplotlib.pylab as plt

p = 1
q = 2
a = 3

x = []
y = []

i = -10
end = 10

while i <= end:
    y.append(i)
    x.append(((i - p)**2)/(4*a) + q)
    i += 0.1

plt.plot(x, y, color='red')
plt.grid()
```
